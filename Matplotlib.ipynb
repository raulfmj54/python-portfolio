# Introducción a Matplotlib

```python
import matplotlib.pyplot as plt, numpy as np
%matplotlib inline

x = np.arange(10)
y = 4 * x + 5 + 10 * np.random.random(size = x.size)
print(x, y, sep='\n\n')
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x, y, marker='o', c='c', s=y*10) #Color: c(permite colores personalizados), Size: s
plt.plot(x, y, linestyle='dashed', color='k')
plt.axis([-1, 10, 0, 50])
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Gráfico xy')
```

## Subplots

```python
fig, ax_arr = plt.subplots(nrows=2, ncols=2)
for ax_fila in ax_arr:
    for ax_columna in ax_fila:
        ax_columna.plot(x, y)
```

## Normal distribution plots

```python
media, std = 10, 1
arr_normal_dist = np.random.normal(media, std, size=10000)
print(arr_normal_dist.mean(), arr_normal_dist.std(), sep='\n')
```

```python
plt.hist(arr_normal_dist)
```

```python
fig, ax = plt.subplots()
freq_arr, bin_arr, _ = ax.hist(arr_normal_dist, facecolor = 'c',
                              alpha = 0.5)
ax.grid(color='#E4E4E4')
ax.set_axisbelow(True)
```

```python
fig, ax = plt.subplots()
ax.boxplot(arr_normal_dist, vert=True, showfliers = True,
          showmeans=True, labels=['Distribución normal'])
```

```python
arr_barras = np.array(['Occidental', 'Central', 'Oriental'])
arr_frec = np.random.randint(0,100,3)
arr_error = np.random.randint(5,100,3)
fig, ax = plt.subplots()
ax.bar(arr_barras, arr_frec, yerr=arr_error, color = 'c')
```

```python
fig, ax_arr = plt.subplots(nrows=2, ncols=2,
                          sharex=False, sharey=False)
fig.set_figwidth(12)
fig.set_figheight(8)
fig.suptitle('Varios gráficos')

#Primer gráfico
g1 = ax_arr[0, 0]
g1.scatter(x, y, marker='o', c='c', s=y*10) #Color: c(permite colores personalizados), Size: s
g1.plot(x, y, linestyle='dashed', color='k')
g1.axis([-1, 10, 0, 50])
g1.set_xlabel('Eje x')
g1.set_ylabel('Eje y')
g1.set_title('Gráfico xy')

#Segundo gráfico
g2 = ax_arr[0, 1]
g2.hist(arr_normal_dist, facecolor='c', alpha= 0.5)
g2.grid(color='#E4E4E4')
g2.set_axisbelow(True)
g2.set_title('Histograma')

#Tercer gráfico
g3 = ax_arr[1, 0]
g3.boxplot(arr_normal_dist, vert=False, showfliers = True,
          showmeans=True, labels=['Distribución normal'])
g3.set_title('Boxplot')

#Cuarto gráfico
g4 = ax_arr[1, 1]
g4.bar(arr_barras, arr_frec, yerr=arr_error, color = 'c')
g4.set_title('Bar plot')
```
