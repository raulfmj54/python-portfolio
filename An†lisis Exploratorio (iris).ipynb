## Visualizaci√≥n de una variable (EDA)

```python
dfIris.hist(bins=15, color='steelblue', edgecolor='black', linewidth=1.0, xlabelsize=8, ylabelsize=8,
           grid=False)
plt.tight_layout(rect=(0, 0, 1.2, 1.2))
```

```python
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.distplot(dfIris['sepal_length'], color='steelblue', ax=axes[0,0])
sns.distplot(dfIris['sepal_width'], color='steelblue', ax=axes[0,1])
sns.distplot(dfIris['petal_length'], color='steelblue', ax=axes[1,0])
sns.distplot(dfIris['petal_width'], color='steelblue', ax=axes[1,1])

```

```python
fig = plt.figure(figsize=(6, 4))
fig.suptitle('Petal_length', fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)
ax=fig.add_subplot(1, 1, 1)
ax.set_xlabel('Petal length')
ax.set_ylabel('Frequency')
freq, bins, patches = ax.hist(dfIris['petal_length'], color='steelblue', bins=15, edgecolor='black', linewidth=1)
```

```python
fig = plt.figure(figsize=(6, 4))
fig.suptitle('Petal_length', fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)
ax=fig.add_subplot(1, 1, 1)
ax.set_xlabel('Petal length')
ax.set_ylabel('Density')
sns.kdeplot(dfIris['petal_length'], color='steelblue',shade=True, ax=ax)
```

```python
sns.kdeplot(dfIris['petal_length'], color='steelblue',shade=True, ax=ax)
```

## An√°lisis multivariable

```python
f, ax = plt.subplots(figsize=(6, 3))
corr = dfIris.corr()
sns.heatmap(round(corr, 2), annot=True, ax=ax, cmap='coolwarm_r', fmt='0.2f', linewidths=0.05, vmin=-1, vmax=1)
f.subplots_adjust(top=0.8)
f.suptitle('Iris daraset: Variables Correlation Heatmap', fontsize=14)
```

```python
dfIris.corr()
```

```python
sns.pairplot(dfIris, hue='species', height=1.8, markers=['o', '+', 'x'], palette='colorblind')
```

```python
g = sns.PairGrid(dfIris, height=1.8)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_level=6)
```

## Coordenadas paralelas

```python
from pandas.plotting import parallel_coordinates
pc = parallel_coordinates(dfIris, 'species', color=('deepskyblue', 'orange', 'forestgreen'))
```

```python
import sklearn.preprocessing as pp
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
scaler = pp.StandardScaler()
scaled_dfIris = scaler.fit_transform(dfIris[cols])
scaled_dfIris = pd.DataFrame(scaled_dfIris, columns=cols)
scaled_dfIris
dfIris_normal = pd.concat([scaled_dfIris, dfIris['species']], axis=1)
dfIris_normal
```

```python
from pandas.plotting import parallel_coordinates
pc = parallel_coordinates(dfIris_normal, 'species', color=('deepskyblue', 'orange', 'forestgreen'))
```
